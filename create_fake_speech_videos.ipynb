Multiple Sentences Cloning

```python
def synth(text,embed):
  # in_fpath = Path("trump10.wav")
  # reprocessed_wav = encoder.preprocess_wav(in_fpath)
  # original_wav, sampling_rate = librosa.load(in_fpath)
  # preprocessed_wav = encoder.preprocess_wav(original_wav, sampling_rate)
  # embed = encoder.embed_utterance(preprocessed_wav)
  with io.capture_output() as captured:
    specs = synthesizer.synthesize_spectrograms([text], [embed])
  generated_wav = vocoder.infer_waveform(specs[0])
  # generated_wav = np.pad(generated_wav, (0, synthesizer.sample_rate), mode="constant")
  return generated_wav
  # generated_wav = np.pad(generated_wav, (0, synthesizer.sample_rate), mode="constant")
  # display(Audio(generated_wav, rate=synthesizer.sample_rate))
```

```python
# sentences = paragraph.split(", ")
# for text in sentences:
#   print(text)
import re
import scipy
sentences = re.split('; |, |\. |\n',text)
# sentences = paragraph.split(" ")
in_fpath = Path(input_audio_path)
reprocessed_wav = encoder.preprocess_wav(in_fpath)
original_wav, sampling_rate = librosa.load(in_fpath)
preprocessed_wav = encoder.preprocess_wav(original_wav, sampling_rate)
embed = encoder.embed_utterance(preprocessed_wav)
generated_wav = None
for sent in sentences:
  print(sent)
  audio = synth(sent,embed)
  # display(Audio(audio, rate=synthesizer.sample_rate))
  if generated_wav is None:
    generated_wav = audio
  else:
    generated_wav = np.append(generated_wav,audio)
    # generated_wav = np.pad(generated_wav, (0, audio), mode="constant")
    # generated_wav = generated_wav + audio
display(Audio(generated_wav, rate=synthesizer.sample_rate))
```

```python
output_audio_path = '/content/drive/MyDrive/create_fake_speech_videos/data/output_data/audio/fake_speech.wav'
write(output_audio_path, synthesizer.sample_rate, generated_wav.astype(np.float32))
```
